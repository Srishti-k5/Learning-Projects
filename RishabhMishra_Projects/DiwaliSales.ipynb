```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Diwali Sales Data.csv", encoding = 'unicode_escape')
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(["Status", "unnamed1"], axis = 1, inplace = True)
df
```

```python
pd.isnull(df)
```

```python
pd.isnull(df).sum()
```

```python
df.shape
```

```python
df.dropna(inplace = True)
df.shape
```

```python
df['Amount'] = df['Amount'].astype('int')
df['Amount'].dtypes
```

```python
df.columns
```

```python
df.rename(columns = {'Marital_Status': 'Shaadi'})
```

```python
df.describe()
```

```python
ax = sns.countplot(x='Gender', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
gender_sales = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = True )
gender_sales
```

```python
sns.barplot(x='Gender', y = 'Amount', data = gender_sales)
```

```python
px = sns.countplot(x='Age Group', hue = 'Gender', data = df)

for bars in px.containers:
    px.bar_label(bars)
```

```python
agegroup_sales = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = True )

sns.barplot(x='Age Group', y = 'Amount', data = agegroup_sales)
```

```python
state_sales = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = True )

sns.set(rc = {'figure.figsize':(25,5)})
sns.barplot(x='State', y = 'Orders', data = state_sales)
```

```python
state_sales = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = True )

sns.set(rc = {'figure.figsize':(25,5)})
sns.barplot(x='State', y = 'Amount', data = state_sales)
```

```python
sns.set(rc = {'figure.figsize':(5,5)})
ax = sns.countplot(x='Marital_Status', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
marital_sales = df.groupby(['Marital_Status', 'Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = True )

sns.set(rc = {'figure.figsize':(10,5)})
sns.barplot(x='Marital_Status', y = 'Amount', data = marital_sales, hue = 'Gender')
```

```python
sns.set(rc = {'figure.figsize':(25,5)})
ax = sns.countplot(x='Occupation', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
occupation_sales = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = True )

sns.set(rc = {'figure.figsize':(25,5)})
sns.barplot(x='Occupation', y = 'Amount', data = occupation_sales)
```

```python
sns.set(rc = {'figure.figsize':(25,5)})
ax = sns.countplot(x='Product_Category', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
product_sales = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = True )

sns.set(rc = {'figure.figsize':(30,5)})
sns.barplot(x='Product_Category', y = 'Amount', data = product_sales)
```

```python
product_sales = df.groupby(['Product_ID'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = True ).head(10)

sns.set(rc = {'figure.figsize':(25,5)})
sns.barplot(x='Product_ID', y = 'Orders', data = product_sales)
```
