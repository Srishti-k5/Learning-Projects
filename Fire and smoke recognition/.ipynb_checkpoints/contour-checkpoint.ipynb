```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt


img = cv2.imread(r"C:\Users\Lenovo\Downloads\WhatsApp Image 2023-02-14 at 17.22.12.jpg")

img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

# lower = np.array([60, 60, 60])
# upper = np.array([250, 250, 250])


# mask = cv2.inRange(img, lower, upper)

thresh = cv2.threshold(img,120,255,cv2.THRESH_BINARY)[1]

contours, hierarchy = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

cont_img = cv2.drawContours(img, contours, -1, 255, 3)

#plt.imshow(cont_img)

# c = max(contours, key = cv2.contourArea)
# x,y,w,h = cv2.boundingRect(c)

#         # draw the biggest contour (c) in green
# cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
# plt.imshow(img)

for c in cont_img:
    x,y,w,h = cv2.boundingRect(c)
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
    plt.imshow(img)
    
    

```
