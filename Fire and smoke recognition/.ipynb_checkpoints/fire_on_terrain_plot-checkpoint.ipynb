```python
import pygmt
```

```python
minlon, maxlon = 75, 110
minlat, maxlat = 20, 45
```

```python
# topo_data = r'C:\Users\Lenovo\.gmt\server\earth\earth_relief\earth_relief_01d_g.grd'
topo_data = '@earth_relief_30s' 
```

```python
fig = pygmt.Figure()
```

```python
# # make color pallets
pygmt.makecpt(
    cmap='topo',
    series='-8000/8000/1000',
    continuous=True
)
```

```python
#plot high res topography
fig.grdimage(
    grid=topo_data,
    region=[minlon, maxlon, minlat, maxlat],
    projection='M4i',
    frame=True
    )
fig.show()
```
