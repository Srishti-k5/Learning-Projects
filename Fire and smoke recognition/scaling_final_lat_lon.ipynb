```python
from PIL import Image
image_path = 'Cambodia Burning.jpeg.jpg'
image = Image.open(image_path)

num_rows = 195
num_cols = 112

cell_width = num_cols/image.width
cell_height = num_rows/image.height

cell_width, cell_height
```

```python
import pandas as pd
lat_lonData = pd.read_excel("coordinates (8).xlsx")
```

```python
# import numpy as np
# data = np.loade('coordinates.txt')
Long = lat_lonData["LONG"];
Lat = lat_lonData["LAT"];
# Long
```

```python
# x1 = cell_width*
grid_cell_val = []
for i in range(len(Long)):
    grid_cell_val.append([Long[i], Lat[i]])

grid_cell_val
# len(grid_cell_val)
```

```python
import pandas as pd
pixelData = pd.read_excel("cambodia_pixel_fire.xlsx")
xCord = pixelData["X-cord"]
yCord = pixelData["Y-cord"]
```

```python
x1 = cell_width*xCord
y1 = cell_width*yCord
```

```python
combined = []
for i in range(len(x1)):
    combined.append(x1[i]*y1[i])
combined = pd.Series(combined)
combined = combined.astype(int)
combined
```

```python
final_lat_long = []
for i in range(len(x1)):
    final_lat_long.append(grid_cell_val[combined[i]])
final_lat_long
```
