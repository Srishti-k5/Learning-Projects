```python
import numpy as np
data = np.loadtxt('india.txt'); #Load data
data #Display a few rows of the dataset
```

```python
import numpy as np
from scipy.interpolate import griddata
```

```python
Long = data[:,0]; Lat = data[:,1]; Elev = data[:,2]; #Variables
pts=1000000
```

```python
[x,y]=np.meshgrid(np.linspace(np.min(Long),np.max(Long),1000),np.linspace(np.min(Lat),np.max(Lat),1000));
z = griddata((Long, Lat), Elev, (x, y), method='linear');
x = np.matrix.flatten(x); #Gridded longitude
y = np.matrix.flatten(y); #Gridded latitude
z = np.matrix.flatten(z); #Gridded elevation
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y,1,z)
plt.colorbar(label='Elevation above sea level [m]')
plt.xlabel('Longitude [Â°]')
plt.ylabel('Latitude [Â°]')
```
