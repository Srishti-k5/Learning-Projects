```python
from PIL import Image
```

```python
img = Image.open('rabbit-breeds-american-white-1553635287.jpg')
image_resized = img.resize((150,150))
```

```python
image_resized.save("rabbit_resized.jpg")
```

```python
#displaying resized image
img_res = mpimg.imread('rabbit_resized.jpg')
img_res_plot = plt.imshow(img_res)
plt.show()
```

# Graying of RGB colored images using CV

```python
pip install opencv-python
```

```python
import cv2
```

```python
cv2.__version__
```

```python
img = cv2.imread("rabbit_resized.jpg")
```

```python
type(img)
```

```python
img.shape
```

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
type(gray_img)
```

```python
gray_img.shape
```

```python
plt.imshow(gray_img)
plt.show()
```

```python
cv2.imwrite("resized_gray_img.jpg", gray_img)
```
